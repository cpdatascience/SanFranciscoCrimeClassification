```python
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing 
import pandas as pd
```

```python
train = pd.read_csv("~/kaggle_data/train.csv")
```

```python
train.head()
```

```python
y = train['Category'].as_matrix()
train.drop('Category',1,inplace=True)
train = pd.get_dummies(train)
train.head()
```

```python
train.columns
```

```python
X = train.as_matrix()
scaler = preprocessing.StandardScaler().fit(X)#scale features
X = scaler.transform(X)
X_train,X_test,y_train,y_test= train_test_split(X,y) # consider using stratified split to account for skewed classes
```

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
```
